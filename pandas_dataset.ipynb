```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/iris.csv"
df = pd.read_csv(url)
print(df.head())

print(df.info())
print(df.describe())

df.plot(kind='box', title='Box Plot of Iris Features', figsize=(10,6))
plt.grid(True)
df.hist()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/iris.csv"
df = pd.read_csv(url)
print(df.head())

print(df.info())
print(df.describe())

df['sepal_length'].plot(kind='hist', title='Sepal Length Distribution', color='red', edgecolor='black')#---->Histogram for sepal length
plt.xlabel('Sepal Length')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/iris.csv"
df = pd.read_csv(url)

plt.scatter(df['sepal_length'], df['petal_length'], color='green',marker='*', label='Sepal Length vs Petal Length')

plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/iris.csv"
df = pd.read_csv(url)

#Line plot for first 10 rows
df.iloc[0:15].plot(kind='line', title='Line Plot of First 10 Samples', marker='o')
plt.grid(True)
plt.show()
```
